```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('WineQT.csv')
```

```python
df = pd.DataFrame(df)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

### The above summary shows how different columns have different range of values. For example, the mean of the column 'volatile acidity' is 0.53 whereas the mean of column 'total sulfur dioxide' is 45.91, which resembles a huge change in values.

```python
df.drop(columns=['Id'],inplace=True)
df.head()
```

```python
df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

cols = df.columns[df.columns != 'quality']

df[cols] = scaler.fit_transform(df[cols])

```

```python
df.head()
```

```python
df.describe()
```

### From the above summary we see that the min and max of all the columns are 0 and 1 respectively. Also the mean and std dev are also in range of 0 to 1.
